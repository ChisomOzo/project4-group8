```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import re
import json
```

```python
# Create a path to csv and read it in pandas dataframe
path=Path("../Resources/movies_metadata.csv")
movies_df=pd.read_csv(path)
movies_df.head()
```

```python
movies_df.info()
```

```python
reduced_movies_df = movies_df.loc[:, ["id","title","genres","poster_path", "overview"]]
reduced_movies_df
```

```python
reduced_movies_df = reduced_movies_df.rename(columns={"id": "movieId"
                                                })
reduced_movies_df.head()
```

```python
# Define a function to clean the genre column
def clean_genre(genre_string):
    # Find all instances of genre names using regular expression
    genres = re.findall(r"'name':\s*'([^']*)'", genre_string)
    # Join the genres with a '|' separator
    return '|'.join(genres)
```

```python
reduced_movies_df = reduced_movies_df.copy()

# Apply the function to the genre column
reduced_movies_df['cleaned_genres'] = reduced_movies_df['genres'].apply(clean_genre)

# Display the DataFrame to verify the results
# If you want to display both original and cleaned genres
print(reduced_movies_df[['cleaned_genres']])
```

```python
non_integer_rows = reduced_movies_df[pd.to_numeric(reduced_movies_df['movieId'], errors='coerce').isna()]
```

```python
reduced_movies_df = reduced_movies_df[pd.to_numeric(reduced_movies_df['movieId'], errors='coerce').notna()]
```

```python
reduced_movies_df['movieId'] = reduced_movies_df['movieId'].astype(int)
```

```python
# Now, drop the 'genres' column from the DataFrame
reduced_movies_df = reduced_movies_df.drop(columns=['genres'])
# Rename cleaned_genres to genres
reduced_movies_df = reduced_movies_df.rename(columns={"cleaned_genres": "genres"})
reduced_movies_df
```

```python
reduced_movies_df['overview'] = reduced_movies_df['overview'].fillna("")
reduced_movies_df['poster_path'] = reduced_movies_df['poster_path'].fillna("")

reduced_movies_df
```

```python
# Transfer data to json

# Convert DataFrame to a list of dictionaries
data_list = reduced_movies_df.to_dict(orient="records")

# Write the list of dictionaries to a JSON file with commas at the end of each line
with open("../Resources/movies_details_cleaned.json", "w") as json_file:
    json_file.write("[\n")
    for i, record in enumerate(data_list):
        json.dump(record, json_file)
        if i < len(data_list) - 1:
            json_file.write(",\n")
    json_file.write("\n]")
```
