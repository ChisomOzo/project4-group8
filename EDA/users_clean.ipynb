```python
# Import Dependencies
import pandas as pd
from pathlib import Path
from io import StringIO
import re
```

```python
# Create a path to csv and read it in pandas dataframe
path = Path("../Resources/users.csv")
users_df = pd.read_csv(path, sep="\t")
users_df.head()
```

```python
# drop unnamed
users_df = users_df.drop("Unnamed: 0", axis=1)
users_df.head()
```

```python
# drop na
users_df = users_df.dropna(how='all')
users_df.head()
```

```python
# check data info
users_df.info()
```

```python
# check counts
users_df.count()
```

```python
users_df.to_csv("../Resources/users_df_cleaned.csv", index=False)
```
